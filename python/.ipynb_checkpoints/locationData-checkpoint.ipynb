```python
#unpack dictionary in dataframe
def unpack(df, column, fillna=None):
    ret = None
    if fillna is None:
        ret = pd.concat([df, pd.DataFrame((d for idx, d in df[column].iteritems()))], axis=1)
        del ret[column]
    else:
        ret = pd.concat([df, pd.DataFrame((d for idx, d in df[column].iteritems())).fillna(fillna)], axis=1)
        del ret[column]
    return ret
```

## Calculate distance

```python
data2 = data.head()
data2[['lat']].shift(-1).fillna(0)
```

```python
data['lat2'] = data[['lat']].shift(-1).fillna(0)
data['lon2'] = data[['lon']].shift(-1).fillna(0)
```

```python
data
```

```python
def haversine(row):
    """
    Calculate the great circle distance between two points
    on the earth (specified in decimal degrees)

    All args must be of equal length.    

    """
    lon1 = row['lon']
    lat1 = row['lat']
    lon2 = row['lon2']
    lat2 = row['lat2']
    
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2

    c = 2 * np.arcsin(np.sqrt(a))
    m = (6367 * c)*1000
    km = 6367 * c
    m = np.round(m, decimals=2)
    return m
```

```python
data['distance_m'] = data.apply(lambda row: haversine(row), axis=1)
```

## calculate speed

```python
data['timedelta'] = (data['datetime']-data['datetime'].shift(-1)).fillna(0)
```

```python
#m/sec
data['timedelta_sec'] = data.timedelta.dt.seconds
```

```python
#set values, where timedifference = 0 to the time it would take with walking speed (2.5m/s)
data.loc[data.timedelta_sec == 0, 'timedelta_sec'] = data.distance_m/2.5
```

```python
data['speed_msec'] = data.distance_m/data.timedelta_sec
data['speed_kmh'] = (data.distance_m/data.timedelta_sec)*3.6
```

## analyze speeds

0-3 kmh: stationary <br>
3-15 kmh: walking or running <br>
15-150 kmh : public transport <br>
150< kmh: flying

```python
def classify_mot(x):
    if x.speed_kmh < 3: return 'stationary'
    elif (x.speed_kmh >= 3) and (x.speed_kmh <15): return 'foot'
    elif (x.speed_kmh >= 15) and (x.speed_kmh <150): return 'vehicle'
    else: return 'plane'
```

```python
data['mot'] = data.apply(classify_mot, axis=1)
```

## subdivide datasets by city

```python
#latmax, latmin, lonmin, lonmax LONDON
maxmin_ld = [51.67339039914003,51.319095575954634,-0.44899886792507004,0.13876968676242996]

#latmax, latmin, lonmin, lonmax SF
maxmin_sf = [37.985007243929694,37.284207361762384,-122.70956609648232,-121.80594060820107]
    
#latmax, latmin, lonmin, lonmax zurich
maxmin_zh = [47.41372315106743,47.33811640957376,8.44761124006584,8.59489670149162]
    
#latmax, latmin, lonmin, lonmax NY
maxmin_ny = [40.943042483075814,40.55790680665638,-74.17674334088838,-73.7551429990915]
    
#latmax, latmin, lonmin, lonmax marrakech
maxmin_mk = [31.72162426370149,31.53266102305224,-8.199414826364318,-7.750140386495332]


def city2(df,coords,name):
    city = df[(coords[0] >= df.lat) & (df.lat >= coords[1]) & (coords[2] <= df.lon)& (df.lon <= coords[3])]
    city = city.reset_index(drop=True)
    #city.to_csv(name +'.csv')
    return city
```

```python
sf =city2(data_red, maxmin_sf, 'san_fran')
lnd =city2(data_red, maxmin_ld, 'london')
zh =city2(data_red, maxmin_zh, 'zh')
mk =city2(data_red, maxmin_mk, 'mk')
ny =city2(data_red, maxmin_ny, 'ny')
```
